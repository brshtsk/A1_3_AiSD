```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt('results.txt', delimiter='\t')
```

```python
n = data[:, 0]
calculated_area = data[:, 1]
```

```python
actual_area = 0.942853
otkl = (calculated_area - actual_area) / actual_area
```

```python
plt.figure(figsize=(10, 5))
plt.plot(n, calculated_area, label='Площадь по Монте-Карло', color='blue')
plt.axhline(y=actual_area, color='red', linestyle='--', label='Реальное значение')
plt.xlabel('n')
plt.ylabel('Площадь')
plt.title('Площадь по Монте-Карло для разных n')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(n, otkl, label='Относительное отклонение по Монте-Карло', color='green')
plt.axhline(y=0, color='red', linestyle='--', label='Реальное значение')
plt.xlabel('n')
plt.ylabel('Относительное отклонение')
plt.title('Относительное отклонение для разных n')
plt.legend()
plt.grid(True)
plt.show()
```
